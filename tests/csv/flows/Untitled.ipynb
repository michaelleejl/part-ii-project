```python
import pandas as pd

flows = pd.read_csv("flows.csv")
flows
```

```python
total_out = flows.groupby("from_city").agg(list)
total_out["total_outflow"] = total_out["volume"].apply(sum)
total_out = total_out.reset_index()
total_out
```

```python
total_in = flows.groupby("to_city").agg(list)
total_in["total_inflow"] = total_in["volume"].apply(sum)
total_in = total_in.reset_index()
```

```python
flows = (
    pd.DataFrame(flows["from_city"])
    .rename({"from_city": "city"}, axis=1)
    .drop_duplicates()
)
res = flows.merge(
    total_out[["from_city", "total_outflow"]].rename({"from_city": "city"}, axis=1),
    on="city",
)
res = res.merge(
    total_in[["to_city", "total_inflow"]].rename({"to_city": "city"}, axis=1), on="city"
)
res
```

```python
res["netflow"] = res["total_outflow"] - res["total_inflow"]
```
